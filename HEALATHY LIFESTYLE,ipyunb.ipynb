```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings 
warnings.filterwarnings('ignore')
```

```python
plt.figure(figsize=(13,8))
sns.heatmap(Healthy_data.corr(),annot=True)
plt.title('Heatmap of Variable Correlations')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(x='Sunshine hours(City)',y='Happiness levels(Country)',data=Healthy_data)
plt.title('Happiness Levels vs Sunshine Hours')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(x='Cost of a monthly gym membership(City)',y='Happiness levels(Country)',data=Healthy_data)
plt.title('Happiness Levels vs Cost of Gym Membership')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(x='Number of take out places(City)',y='Happiness levels(Country)',data=Healthy_data)
plt.title('Happiness Levels vs Number of Takeouts')
plt.show
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(x='Annual avg. hours worked',y='Happiness levels(Country)',data=Healthy_data)
plt.title('Happiness Levels vs Annual Average Hours Worked')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(x='Outdoor activities(City)',y='Happiness levels(Country)',data=Healthy_data)
plt.title('Happiness Levels vs Outdoor Activities')
plt.show()
```
