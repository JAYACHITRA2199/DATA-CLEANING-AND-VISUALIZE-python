# DATA VISUALITION

```python
plt.figure(figsize=(15,6))
sns.countplot(diamonds_data['color'],data=diamonds_data,palette='hls')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.countplot(diamonds_data['cut'],data=diamonds_data,palette='hls')
plt.xticks(rotation=90)
plt.show()
```

```python
diamonds_data=diamonds_data.data.drop(diamonds_data[diamonds_data["x"]==0)].index)
diamonds_data=diamonds_data.drop(diamonds_data[diamonds_data["y"]==0)].index)
diamonds_data=diamonds_data.drop(diamonds_data[diamonds_data["z"]==0)].index)

```

```python
diamonds_data.shape
```

```python
sns.pairplot(diamonds_data)
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.violinplot(x='cut',y='price',data=diamonds_data)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.violinplot(x='clarity',y='price',data=diamonds_data)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.countplot(diamonds_data['clarity'],data=diamonds_data,palette='hls')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.violinplot(x='carat',y='price',z='depth',data=diamonds_data)
plt.xticks(rotation=90)
plt.show()
```
