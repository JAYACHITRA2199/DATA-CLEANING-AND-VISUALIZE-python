```python
feature_na=[feature for feature in universtity_data.columns if universtity_data[feature].isnull().sum()>0]
feature_na
```

```python
for feature in feature_na:
    print('{} has {} % missing values'.format(feature,np.round(universtity_data[feature].isnull().sum()/len(universtity_data)*100,4)))
```

```python
df3 = pd.DataFrame()
df3['Count of the universites'] = universtity_data.groupby('location')['location'].count().sort_values(ascending=False)
df3 = df3.reset_index()
df3['location code'] = universtity_data['location code']
df3
```

```python
plt.figure(figsize=(30,15))
plt.xticks(rotation=90)
plt.title('Number of universities included from the country in world rankings')
ax = sns.barplot(x='location code', y= 'Count of the universites',data=df3)
ax.bar_label(ax.containers[0])
```

```python
plt.figure(figsize=(18,6))
sns.countplot(universtity_data['location'],data=universtity_data,palette='hls')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.distplot(universtity_data['Rank'])
plt.show()
```
